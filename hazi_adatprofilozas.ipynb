```sql
-- Leiras NULL értékei

SELECT sor_tipus, COUNT(*) AS 'DB'
FROM
(SELECT
    CASE WHEN Leiras IS NULL THEN 'NULL érték'
        ELSE 'Nem NULL'
        END AS sor_tipus
        FROM Termek
)ROWS
GROUP BY sor_tipus
```

```sql
SELECT készleten, COUNT(*) AS 'DB'
FROM 
    (SELECT
        CASE
        WHEN Keszlet < 500 THEN 'Kevesebb mint 500'
        WHEN Keszlet BETWEEN 500 AND 2000  THEN '500-2000 között'
        ELSE 'Több mint 2000'
         END AS készleten
       FROM Termek
    ) rows
GROUP BY készleten
```

```sql
SELECT MIN(Listaar) as 'Legkisebb megadott listaár',
        MAX(Listaar) as 'Legnagyobb megadott listaár',
        AVG(Listaar) as 'Átlagos listaár',
        STDEV(Listaar) as 'Listaár szórása',
        VAR(Listaar) as 'Variancia'
FROM Termek
```
