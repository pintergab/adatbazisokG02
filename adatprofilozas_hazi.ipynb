```sql
--1.feladat : NULL értékű szállítási címek száma


SELECT sor_tipus, COUNT(*) AS 'DB'
FROM 
    (SELECT 
        CASE WHEN SZALL_CIM IS NULL THEN 'NULL érték'
            ELSE 'Nem NULL'
            END AS sor_tipus
            FROM Rendeles
    ) ROWS
GROUP BY sor_tipus
```

```sql
--2. feladat : Korcsoportok szerinti csoportosítás

SELECT  kor,COUNT(*) as'Ügyfelek száma'
FROM
    (SELECT     
        CASE 
            WHEN SZULEV BETWEEN 1965 AND 1975 THEN '50-60'
            WHEN SZULEV BETWEEN 1975 AND 1985 THEN '40-50'
            ELSE '<40'
            END AS kor
            FROM Ugyfel) rows

GROUP BY kor
        
```

```sql
SELECT MIN(MENNYISEG) as 'Legkevesebb mennyiség',
        MAX(MENNYISEG) as 'Legtöbb mennyiség',
        AVG(MENNYISEG) as 'Átlagos mennyiség',
        STDEV(MENNYISEG) as 'Mennyiségek szórása',
        VAR(MENNYISEG) as 'Variancia'
FROM Rendeles_tetel
```
